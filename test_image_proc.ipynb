```python
import pytesseract
from PIL import Image
import os
import ocr_processing 
import importlib
import matplotlib.pyplot as plt

# Remplacez le chemin par le chemin absolu de votre dossier "error"
folder_path = "error"
param = '--psm 6 -c tessedit_char_whitelist=0123456789-=+Ã· tessedit_char_blacklist=/'
# Fonction pour traiter chaque image et extraire le texte
def process_images(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith(".png") or filename.endswith(".jpg") or filename.endswith(".jpeg"):
            image_path = os.path.join(folder_path, filename)
            try:
                image = Image.open(image_path)
                tresh = ocr_processing.image_processing(image)
                print(pytesseract.image_to_string(tresh, config=param))
                display(image)
                plt.imshow(tresh)
                text = pytesseract.image_to_string(image, config=param)
                print(f"Texte extrait de {filename} :\n{text}\n")
            except Exception as e:
                print(f"Erreur lors du traitement de {filename}: {e}")

# ExÃ©cute la fonction
process_images(folder_path)
```

```python
import cv2
import pytesseract

folder_path = "error"
image_name = os.listdir(folder_path)[1]
image_path = os.path.join(folder_path, image_name)
image = cv2.imread(image_path)
if image is None:
    print(f"Failed to load image: {image_path}")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

custom_config = r'--oem 3 --psm 6' 

text = pytesseract.image_to_string(thresh, config=custom_config)
print(text)
plt.imshow(image)
```
